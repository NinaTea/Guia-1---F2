Me piden estudiar el movimiento de las masas para el caso estacionario y ver cuándo están en fase o en contrafase. El movimiento de las pertículas va a responder a estas ecuaciones:

$\varphi_a (t)$ = $(A^{(1)}_a-A^{(2)}_a)sen(\Omega t)$ + $(A^{(1)}_e-A^{(2)}_e)cos(\Omega t)$


 $\varphi_b (t)$ = $(A^{(1)}_a+A^{(2)}_a)sen(\Omega t)$ + $(A^{(1)}_e+A^{(2)}_e)cos(\Omega t)$
Donde  $A^{(i)}_j$  i= 1,2 y j= a, e corresponden a la amplitud absorbente 1 y 2, y a la amplitud elástica 1 y 2.

Sus expresiones son:

 $A^{(1)}_a = \frac{F_o}{2m}\frac{\Gamma \Omega }{\Gamma^{2} \Omega^{2} +(\omega_1^{2}-\Omega^{2} )^{2}}$
  $A^{(2)}_a = \frac{F_o}{2m}\frac{\Gamma \Omega }{\Gamma^{2} \Omega^{2} +(\omega_2^{2}-\Omega^{2} )^{2}}$



   $A^{(1)}_e = \frac{F_o}{2m}\frac{(\omega_1^{2}-\Omega^{2}) }{\Gamma^{2} \Omega^{2} +(\omega_1^{2}-\Omega^{2} )^{2}}$
   $A^{(2)}_e = \frac{F_o}{2m}\frac{(\omega_2^{2}-\Omega^{2}) }{\Gamma^{2} \Omega^{2} +(\omega_2^{2}-\Omega^{2} )^{2}}$



$\omega_1^{2} = \frac{g}{l}$ 
$\omega_2^{2} = \frac{g}{l}+ \frac{2k}{m}$




```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as iwy  
```

```python
m = 1 #kg
gamma = 0.1 #s**(-1)
Fo = 1 #N
w_1 = 2 
w_2 = 4
omegon = np.linspace(0,8,500) 

A_abs_1 = (Fo/2*m)*gamma*omegon / (gamma**2 * omegon**2 + (w_1**2 - omegon**2)**2 )
A_abs_2 = (Fo/2*m)*gamma*omegon / (gamma**2 * omegon**2 + (w_2**2 - omegon**2)**2 )

A_el_1 = (Fo/2*m) * ( (w_1**2 - omegon**2) / ((gamma * omegon)**2 + (w_1**2 - omegon**2)**2) )
A_el_2 = (Fo/2*m) * ( (w_2**2 - omegon**2) / ((gamma * omegon)**2 + (w_2**2 - omegon**2)**2) )

A_abs_a = np.add(A_abs_1, -1*A_abs_2)
A_abs_b = np.add(A_abs_1, A_abs_2)

A_el_a = np.add(A_el_1, -1*A_el_2)
A_el_b = np.add(A_el_1, A_el_2)

plt.figure(figsize=(10,6))
plt.plot(omegon, A_abs_a, "darkviolet", label="A_abs_a")
plt.plot(omegon, A_el_a, "deeppink", label="A_el_a")
plt.plot([w_1, w_1], (-1.5,2.5), "--", color = "green")
plt.text(w_1 + 0.2, 2.4, "$\omega_1$",color = "green")
plt.plot([w_2, w_2], (-1.5,2.5), "--", color = "green")
plt.text(w_2 + 0.2, 2.4, "$\omega_2$",color = "green")
plt.xlabel("$\Omega(t)$", size = 14)
plt.ylabel("Amplitudes", size = 14)
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.plot(omegon, A_abs_b, "brown", label="A_abs_b")
plt.plot(omegon, A_el_b, "red", label="A_el_b")
plt.plot([w_1, w_1], (-1.5,2.5), "--", color = "b")
plt.text(w_1 + 0.2, 2.4, "$\omega_1$",color = "b")
plt.plot([w_2, w_2], (-1.5,2.5), "--", color = "b")
plt.text(w_2 + 0.2, 2.4, "$\omega_2$",color = "b")
plt.xlabel("$\Omega(t)$", size = 14)
plt.ylabel("Amplitudes", size = 14)
plt.legend()
plt.grid()
```

Mirando ambos gráficos, vemos cómo para $\Omega$ = $\omega_1$ las masas se mueven en fase y para $\Omega$ = $\omega_2$ en contrafase.

En el inciso e me piden graficar $\frac{\Psi_a}{\Psi_b}$ para ver cuánto del movimiento de la masa a se transfiere a la masa b.

La relación está dada por:


*   $\frac{\Psi_a}{\Psi_b} = \frac{\omega_2^{2} - \omega_1^{2}}{\omega_2^{2} + \omega_1^{2} - 2\Omega^{2}}$

```python
psi_a = w_2**2 - w_1**2
psi_b = w_2**2 + w_1**2 -2*(omegon**2)

transferencia = psi_a/psi_b

plt.figure(figsize=(10,6))
plt.scatter(omegon, transferencia, color = "r", label = "psi_a/psi_b", s=1.5)
plt.plot([2, 2], (-2.0,2.0), "--", color = "b")
plt.plot([4, 4], (-2.0,2.0), "--", color = "b")
plt.plot([0, 8], (0,0), "--", color = "b")
plt.text(0.0, 0.20, "$\Omega$",color = "b")
plt.ylim(-2,2)
plt.legend()
plt.grid()
```

Ahora podemos ver casos particulares que nos interesen

```python
def distintos_casos(Omegon = 1.0, Gamma=0.1):
  t = np.linspace(0,25,250)
  omegon2 = Omegon
  A_abs_1 = (Fo/2*m)*Gamma*omegon2 / ((Gamma)**2 * omegon2**2 + (w_1**2 - omegon2**2)**2 )
  A_abs_2 = (Fo/2*m)*Gamma*omegon2 / (Gamma**2 * omegon2**2 + (w_2**2 - omegon2**2)**2 )

  A_el_1 = (Fo/2*m) * ( (w_1**2 - omegon2**2) / ((Gamma * omegon2)**2 + (w_1**2 - omegon2**2)**2) )
  A_el_2 = (Fo/2*m) * ( (w_2**2 - omegon2**2) / ((Gamma * omegon2)**2 + (w_2**2 - omegon2**2)**2) )

  A_abs_a = np.add(A_abs_1, -1*A_abs_2)
  A_abs_b = np.add(A_abs_1, A_abs_2)

  A_el_a = np.add(A_el_1, -1*A_el_2)
  A_el_b = np.add(A_el_1, A_el_2)

  psi_a_ec = A_abs_a*np.sin(omegon2*t) + A_el_a*np.cos(omegon2*t)
  psi_b_ec = A_abs_b*np.sin(omegon2*t) + A_el_b*np.cos(omegon2*t)

  plt.figure(figsize=(10,6))
  plt.title("Posiciones en función del tiempo")
  plt.plot(t, psi_a_ec,"r", label = "psi_a")
  plt.plot(t, psi_b_ec,"b", label = "psi_b")
  plt.xlabel("t[s]")
  plt.ylabel("Amplitudes")
  plt.legend()
  plt.ylim(-3,3)
  plt.grid()
iwy.interact(distintos_casos, Omegon = (0,5,0.1), Gamma = (0,1,0.01));
```
